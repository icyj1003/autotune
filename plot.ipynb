```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
```

```python
y, sr = librosa.load('raw.wav')
f0, voiced_flag, voiced_probs = librosa.pyin(y,
                                             fmin=librosa.note_to_hz('C2'),
                                             fmax=librosa.note_to_hz('C7'))
times = librosa.times_like(f0)
```

```python
y1, sr1 = librosa.load('./nangtho/vocals.wav')
f0_1, voiced_flag_1, voiced_probs_1 = librosa.pyin(y1,
                                             fmin=librosa.note_to_hz('C2'),
                                             fmax=librosa.note_to_hz('C7'))
times_1 = librosa.times_like(f0_1)
```

```python
y2, sr2 = librosa.load('out.wav')
f0_2, voiced_flag_2, voiced_probs_2 = librosa.pyin(y2,
                                             fmin=librosa.note_to_hz('C2'),
                                             fmax=librosa.note_to_hz('C7'))
times_2 = librosa.times_like(f0_2)
```

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(times, f0, label='raw', color='red', linewidth=2)
ax.plot(times_1, f0_1, label='original', color='black', linewidth=4)
ax.plot(times_2, f0_2, label='tunned', color='green', linewidth=2)
ax.legend(loc='upper right', prop={'size': 30})
ax.set_xlabel('time', fontsize=30)
ax.set_ylabel('Hz', fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
